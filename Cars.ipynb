```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('cars.csv')
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:, -1].values
print(X)
print(Y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
CategoricalColumns=['Car','Model']
#Getting Dummy Pandas
# df= pd.get_dummies(dataset, columns=CategoricalColumns)
# ohe=OneHotEncoder()
# transformed_data= ohe.fit_transform(dataset[CategoricalColumns])
# encoded_data = pd.DataFrame(transformed_data, index=dataset.index)
# concatenated_data=pd.concat([dataset, encoded_data], axis=1)

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
ct.fit_transform(X)
X=np.array(ct.fit_transform(X))
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
print(X_train.shape)
print(Y_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
Y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((Y_pred.reshape(len(Y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```
